```python
import mglearn
import matplotlib.pyplot as plt

# データ生成
X, y  = mglearn.datasets.make_forge()

print(X) # 2次元配列の
print(X[:, 0]) # 1番目の要素のみ抽出
print(X[:, 1]) # 2番目の要素のみ抽出
print(y)
# データのプロット
#discrete_scatterは第三引数に各データに対するラベルを指定することにより、
#そのラベルに応じてMarkerの種類や色を変えて散布図を描写をしてくれる
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(['Class 0', 'Class 1'], loc=4)
plt.xlabel('First Feature')
plt.ylabel('Secound Feature')
```

```python
import mglearn
import matplotlib.pyplot as plt

X , y = mglearn.datasets.make_wave(n_samples=40)

plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel("Feature")
plt.ylabel("target")
```
