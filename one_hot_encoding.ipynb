```python
# ダミー変数を使う
# ダミー変数はカテゴリ変数を1つ以上の0と1の値をもつ新しい特徴量に置き換えるものである。
# 0と1を使うことで、線形2クラス分類の式を当てはめることができる
import pandas as pd
import mglearn
import os



adult_path = os.path.join(mglearn.datasets.DATA_PATH, 'adult.data')
data = pd.read_csv(
    adult_path, header=None, index_col=False, names=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'capotal-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']
)
data.head()
```

```python
print(data.gender.value_counts())
```

```python
print(f"Original features: {list(data.columns)}")
data_dummies = pd.get_dummies(data)
print(f"Features after get_dummies {list(data_dummies.columns)}")
data_dummies.head()
```

```python
features = data_dummies.loc[:, 'age': 'occupation_ Transport-moving']
# 処理させるためにnumpy行列を取り出す
X = features.values
y = data_dummies['income_ >50K'].values
print(f"X.shape: {X.shape} y.shape: {y.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
print(f"Test score: {round(logreg.score(X_test, y_test), 3)}")
```
